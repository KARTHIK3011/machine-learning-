```
import matplotlib.pyplot as plt

x=[1,10]
y=[2,20]
plt.plot(x,y, marker='*', c = 'r')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show
```

```
x=[1,10,34,45]
y=[2,20,56,23]
plt.plot(x,y, marker='D', linestyle= ':', c = 'k')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line plot")
plt.show
```
